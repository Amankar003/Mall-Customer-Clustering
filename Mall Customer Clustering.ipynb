```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Mall_Customers.csv')
```

```python
data.head()
```

```python
x = data.iloc [:,[3,4]].values
```

```python
x.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x = sc.fit_transform(x)
```

```python
x.shape
```

```python
from sklearn.cluster import KMeans  # Ensure you import KMeans correctly
wcss = []  # Initialize an empty list to store WCSS values

for i in range(1, 11):
    kmeans_model = KMeans(n_clusters=i, init="k-means++", random_state=0)  # Avoid using KMeans as a variable name
    kmeans_model.fit(x)  # Fit the model to your data
    wcss.append(kmeans_model.inertia_)  # Append the inertia (WCSS) value

# You can plot the WCSS values to find the optimal number of clusters
import matplotlib.pyplot as plt
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans  # Ensure you import KMeans correctly

# Initialize your KMeans model and fit it to data
kmeans_model = KMeans(n_clusters=5, random_state=0)  # Correct usage of the KMeans class
kmeans_model.fit(x)  # x is your data
y_cluster = kmeans_model.predict(x)  # Predict cluster labels

# You can print or use the predicted labels y_cluster
print(y_cluster)
```
